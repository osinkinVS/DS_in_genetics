import pandas as pd
from Bio import Entrez
from Bio import SeqIO
import xml.etree.ElementTree as ET

**Укажите ваш email для NCBI**

Entrez.email = "your_email@example.com"

def fetch_and_create_table(gene_id):

    Функция принимает идентификатор гена,
    собирает информацию и создает новую запись в таблице.
    
    # Запрашиваем данные по гену
    handle = Entrez.efetch(db="gene", id=gene_id, rettype='xml')
    gene_xml = handle.read()
    handle.close()
    
    # Парсим XML файл
    root = ET.fromstring(gene_xml)
    gene_locus = root.find(".//Entrezgene/Entrezgene_locus")
    region = gene_locus.find(".//Gene-commentary/Gene-commentary_seqs/Seq-loc/Seq-loc_int/Seq-interval")
    
    start = int(region.find("Seq-interval_from").text) + 1
    end = int(region.find("Seq-interval_to").text) + 1
    gi_id = region.find("Seq-interval_id/Seq-id/Seq-id_gi").text
    strand = region.find("Seq-interval_strand/Na-strand").get("value")
    
    # Загружаем соответствующую хромосому
    handle = Entrez.efetch(db="nucleotide", id=gi_id, rettype='fasta', retmode='text')
    chromosome = SeqIO.read(handle, 'fasta')
    handle.close()
    
    # Выбираем участок генома, соответствующий данному гену
    sequence = str(chromosome.seq[start:end])
    
    # Создаем динамически словарь
    data_dict = {
        'Ген_ID': gene_id,
        'Название_болезни': '',  # Предполагается, что имя болезни известно отдельно
        'Хромосома': chromosome.name.split('|')[1].strip(),  # Имя хромосомы
        'Начало': start,
        'Конец': end,
        'Стрнд': strand,
        'Последовательность': sequence[:50] + '...' if len(sequence) > 50 else sequence
    }
    
    return data_dict

## Для полученного гена найдем аминокислотную последовательность, количество транскриптов, количество аминокислот для самой длинной изоформы белка

```python
from Bio.SeqUtils import seq1
from Bio.Seq import reverse_complement
if strand == 'minus':
    # Обратная цепочка, делаем reverse complement
    geneACE2 = reverse_complement(chromosome.seq[start:end])
else:
    # Прямая цепочка, оставляем последовательность как есть
    geneACE2 = chromosome.seq[start:end]

protein_seq = str(seq1(str(geneACE2)))
```

```python
geneACE2
```

```python
protein = geneACE2.translate(to_stop=True)
protein
```

```python
longest_protein_length = len(protein)
```

```python
transcripts = 7 # посмотрели на сервисе
```

```python
df["protein"] = str(protein)
df["num_transcripts"] = transcripts
df["longest_protein_length"] = longest_protein_length
```
