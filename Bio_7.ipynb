Исследуется влияние меди на устойчивость к антибиотикам у резистентных и чувствительных бактерий P.aeruginosa

```python
!wget https://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/006/765/GCF_000006765.1_ASM676v1/GCF_000006765.1_ASM676v1_genomic.fna.gz
!gunzip GCF_000006765.1_ASM676v1_genomic.fna.gz
!mv GCF_000006765.1_ASM676v1_genomic.fna PAO1_ref.fa
```

```python
!fastqc SRR12905311_1.fastq SRR12905311_2.fastq
```

```python
!java -jar Trimmomatic-0.39/trimmomatic-0.39.jar PE -phred33 \
SRR12905311_1.fastq SRR12905311_2.fastq \
SRR12905311_1_trimmed.fastq SRR12905311_1_unpaired.fastq \
SRR12905311_2_trimmed.fastq SRR12905311_2_unpaired.fastq \
ILLUMINACLIP:TruSeq3-PE.fa:2:30:10 LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36
```

```python
!bwa index PAO1_ref.fa
!bwa mem -t 2 PAO1_ref.fa SRR12905311_1_trimmed.fastq SRR12905311_2_trimmed.fastq > aligned.sam
```

```python
!samtools view -bS aligned.sam > aligned.bam
!samtools sort aligned.bam -o aligned_sorted.bam
!samtools index aligned_sorted.bam
```

```python
!samtools faidx PAO1_ref.fa
!bcftools mpileup -Ou -f PAO1_ref.fa aligned_sorted.bam | \
bcftools call -vmO z --ploidy 1 -o variants.vcf.gz
!bcftools index variants.vcf.gz
```

```python
!bcftools view -v snps variants.vcf.gz -i '%QUAL>=20' -Oz -o filtered_snps.vcf.gz
!bcftools index filtered_snps.vcf.gz
```

```python
!bcftools stats filtered_snps.vcf.gz > stats.txt
!head -n 20 stats.txt
```

Высокий показатель Ts/Tv ratio (~3.24):Это свидетельствует о значительном преобладании транзиций над трансверсиями. Такой высокий коэффициент является нормальным для бактериальных геномов и указывает на высокую степень консервативности изменений.  
Доминирование определенных транзиций:Преобладающие типы транзиций включают замену аденина на гуанин (A→G) и тимина на цитозин (T→C). Эти изменения отражают стабильный характер эволюции бактериальной ДНК, поскольку такие замены менее склонны вызывать значительные функциональные нарушения.  
Редкость трансверсий:Трансверсии, такие как A→C и T→G, наблюдаются значительно реже. Такие замены чаще вызывают существенные изменения в структуре белка, что делает их редкими вследствие негативного отбора.  
Значение высокой доли транзиций:Высокая доля транзиций связана с меньшим количеством вредных эффектов на белки и геномы, позволяя бактериям сохранять стабильность своего генома. Таким образом, этот феномен отражает адаптивную стратегию бактерий, минимизирующую риск негативных последствий мутаций.​

Анализ данных показывает значительную долю транзиций (особенно A→G и T→C), что характерно для большинства бактериальных геномов. Высокий Ts/Tv ratio подтверждает тенденцию к сохранению стабильности генома, несмотря на происходящие мутации. Этот механизм обеспечивает эффективную адаптацию бактерий к различным условиям окружающей среды.
