```python
# Распаковка FASTQ
!gunzip -k SRR5365416.fastq.gz
```

```python
# Шаг 1: Проверка качества ридов
!fastqc SRR5365416.fastq
```

```python
# Шаг 2: Триммирование ридов (single-end)
!java -jar /content/Trimmomatic-0.39/trimmomatic-0.39.jar SE \
    SRR5365416.fastq \
    SRR5365416_trimmed.fastq \
    ILLUMINACLIP:/content/Trimmomatic-0.39/adapters/TruSeq3-SE.fa:2:30:10 \
    LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36
```

```python
# Шаг 3: Индексирование референсного генома
!bwa index Cl_perfringens.fa
```

```python
# Шаг 4: Выравнивание ридов
!bwa mem Cl_perfringens.fa SRR5365416_trimmed.fastq > aligned.sam
```

```python
# Шаг 5: Конвертация и сортировка BAM
!samtools view -S -b aligned.sam > aligned.bam
!samtools sort aligned.bam -o aligned_sorted.bam
!samtools index aligned_sorted.bam
```

```python
# Шаг 6: Индексирование референсного генома
!samtools faidx Cl_perfringens.fa
```

```python
# Шаг 7: Вызов вариантов
!bcftools mpileup -f Cl_perfringens.fa aligned_sorted.bam | bcftools call -mv -Ob -o variants.bcf
!bcftools index variants.bcf
```

```python
# Шаг 8: Фильтрация вариантов
!bcftools view variants.bcf > variants.vcf
!bcftools filter -i 'QUAL>20 && DP>10' variants.vcf > filtered_variants.vcf
```

```python
# Шаг 9: Аннотация вариантов (упрощенная версия)
# Создаем BED файл с генами из GTF
!awk 'BEGIN{OFS="\t"} $3=="gene" {split($10,a,"\""); print $1,$4,$5,a[2]}' Cl_perfringens.gtf > genes.bed
```

```python
# Аннотируем варианты с помощью bedtools
!bedtools intersect -a filtered_variants.vcf -b genes.bed -wa -wb > annotated_variants.tsv
```

```python
# Шаг 10: Генерация отчета
!multiqc . -o multiqc_report
```
